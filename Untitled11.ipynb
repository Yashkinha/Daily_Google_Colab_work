````python
from collections import defaultdict

visited = defaultdict(lambda: False)
j1 = 0
j2 = 0
l = 0

def water_jug(X, Y):
    global j1, j2, l

    if (X == l and Y == 0) or (X == 0 and Y == l):
        print("(", X, ",", Y, ")", sep="")
        return True

    if not visited[(X, Y)]:
        print("(", X, ",", Y, ")", sep="")

        visited[(X, Y)] = True

        return (
            water_jug(0, Y) or
            water_jug(X, 0) or
            water_jug(j1, Y) or
            water_jug(X, j2) or
            water_jug(X - min(X, (j2 - Y)), Y + min(X, (j2 - Y))) or
            water_jug(X + min(Y, (j1 - X)), Y - min(Y, (j1 - X)))
        )
    else:
        return False

j1 = int(input("Enter capacity of j1: "))
j2 = int(input("Enter capacity of j2: "))
l = int(input("Enter liter: "))
water_jug(0, 0)
````

````python
from collections import deque

def BFS(a, b, target):
    m = {}
    isSolvable = False
    path = []

    q = deque()

    q.append((0, 0))

    while len(q) > 0:
        u = q.popleft()

        if (u[0], u[1]) in m:
            continue

        if u[0] > a or u[1] > b or u[0] < 0 or u[1] < 0:
            continue

        path.append([u[0], u[1]])

        m[(u[0], u[1])] = 1

        if u[0] == target or u[1] == target:
            isSolvable = True

            if u[0] == target and u[1] != 0:
                path.append([u[0], 0])
            elif u[0] != 0:
                path.append([0, u[1]])

            for i in range(len(path)):
                print("(", path[i][0], ",", path[i][1], ")")
            break

        q.append([u[0], b])
        q.append([a, u[1]])

        for ap in range(max(a, b) + 1):
            c = u[0] + ap
            d = u[1] - ap
            if c == a or (d == 0 and d >= 0):
                q.append([c, d])

            c = u[0] - ap
            d = u[1] + ap
            if (c == 0 and c >= 0) or d == b:
                q.append([c, d])

        q.append([a, 0])
        q.append([0, b])

    if not isSolvable:
        print("No solution")

if __name__ == '__main__':
    Jug1 = int(input("Enter size of 1st jug: "))
    Jug2 = int(input("Enter the size of 2nd jug: "))
    target = int(input("Enter the size of target jug: "))
    print("\nPath from initial state to solution state:")
    BFS(Jug1, Jug2, target)
````
