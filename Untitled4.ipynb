```python
import time
import random

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def average_bubble_sort(arr):
    start = time.time()
    bubble_sort(arr)
    finish = time.time()
    return finish - start

def best_bubble_sort(n):
    arr = list(range(n))
    start = time.time()
    bubble_sort(arr)
    finish = time.time()
    return finish - start

def worst_bubble_sort(n):
    arr = list(range(n, 0, -1))
    start = time.time()
    bubble_sort(arr)
    finish = time.time()
    return finish - start

if __name__ == "__main__":
    T = int(input("Enter the number of values of n: "))
    for k in range(T):
        n = int(input("Enter the value of n: "))
        lb = 0
        rb = 100
        arr = [random.randint(lb, rb) for _ in range(n)]
        print()
        best_time = best_bubble_sort(n)
        print(f"Time Taken for best case = {best_time * 1e6} microseconds")
        average_time = average_bubble_sort(arr.copy())
        print(f"Time Taken for average case = {average_time * 1e6} microseconds")
        worst_time = worst_bubble_sort(n)
        print(f"Time Taken for worst case = {worst_time * 1e6} microseconds")
        print()
```
