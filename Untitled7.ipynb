```python
class Book:
    def __init__(self):
        self.books = []

    def add_book(self, name, author):
        book = {
            'name': name,
            'author': author
        }
        self.books.append(book)
        print(f"Book '{name}' added ")

    def search_book(self, name):
        for book in self.books:
            if book['name'].lower() == name.lower():
                print(f"Book Found!\nName: {book['name']}\nAuthor: {book['author']}")
                return
        print(f"Book with name '{name}' not found.")

bookshop = Book()
bookshop.add_book("Python_Book", "Yash")
bookshop.add_book("DAA_Book", "Har")
bookshop.add_book("AI_Book", "Me")
bookshop.search_book("Python_Book")
bookshop.search_book("D")
bookshop.search_book("AI_Book")
```
