```python
import random
import math
class RSA:
    def __init__(self, key_size=250):
        self.prime = set()
        self.public_key = None
        self.private_key = None
        self.n = None
        self.key_size = key_size
    def primefiller(self):
        sieve = [True] * self.key_size
        sieve[0] = False
        sieve[1] = False
        for i in range(2, self.key_size):
            for j in range(i * 2, self.key_size, i):
                sieve[j] = False
        for i in range(len(sieve)):
            if sieve[i]:
                self.prime.add(i)
    def pickrandomprime(self):
        k = random.randint(0, len(self.prime) - 1)
        it = iter(self.prime)
        for _ in range(k):
            next(it)
        ret = next(it)
        self.prime.remove(ret)
        return ret

    def setkeys(self):
        prime1 = self.pickrandomprime()
        prime2 = self.pickrandomprime()
        self.n = prime1 * prime2
        fi = (prime1 - 1) * (prime2 - 1)
        e = 2
        while True:
            if math.gcd(e, fi) == 1:
                break
            e += 1
        self.public_key = e
        d = 2
        while True:
            if (d * e) % fi == 1:
                break
            d += 1
        self.private_key = d
    def encrypt(self, message):
        e = self.public_key
        encrypted_text = 1
        while e > 0:
            encrypted_text *= message
            encrypted_text %= self.n
            e -= 1
        return encrypted_text
    def decrypt(self, encrypted_text):
        d = self.private_key
        decrypted = 1
        while d > 0:
            decrypted *= encrypted_text
            decrypted %= self.n
            d -= 1
        return decrypted
    def encoder(self, message):
        encoded = []
        for letter in message:
            encoded.append(self.encrypt(ord(letter)))
        return encoded
    def decoder(self, encoded):
        s = ''
        for num in encoded:
            s += chr(self.decrypt(num))
        return s

if __name__ == '__main__':
    rsa = RSA()
    rsa.primefiller()
    rsa.setkeys()
    message = input("Enter the message: ")
    coded = rsa.encoder(message)
    print("\nInitial message:",message)
    print("\nThe encoded message (encrypted by public key):")
    print(' '.join(map(str, coded)))
    print("\nThe decoded message (decrypted by public key):")
    print(rsa.decoder(coded))
```
