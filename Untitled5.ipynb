```python
import time
import random

def merge(arr, left, mid, right):
    n1 = mid - left + 1
    n2 = right - mid

    left_arr = [0] * n1
    right_arr = [0] * n2

    for i in range(n1):
        left_arr[i] = arr[left + i]

    for j in range(n2):
        right_arr[j] = arr[mid + 1 + j]

    i = 0
    j = 0
    k = left

    while i < n1 and j < n2:
        if left_arr[i] <= right_arr[j]:
            arr[k] = left_arr[i]
            i += 1
        else:
            arr[k] = right_arr[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = left_arr[i]
        i += 1
        k += 1

    while j < n2:
        arr[k] = right_arr[j]
        j += 1
        k += 1

def merge_sort(arr, left, right):
    if left < right:
        mid = (left + right) // 2
        merge_sort(arr, left, mid)
        merge_sort(arr, mid + 1, right)
        merge(arr, left, mid, right)

def average_merge_sort(arr):
    start = time.time()
    merge_sort(arr, 0, len(arr) - 1)
    finish = time.time()
    return finish - start

def best_merge_sort(n):
    arr = list(range(n))
    start = time.time()
    merge_sort(arr, 0, len(arr) - 1)
    finish = time.time()
    return finish - start

def worst_merge_sort(n):
    arr = list(range(n, 0, -1))
    start = time.time()
    merge_sort(arr, 0, len(arr) - 1)
    finish = time.time()
    return finish - start

if __name__ == "__main__":
    T = int(input("Enter the number of values of n: "))
    for k in range(T):
        n = int(input("Enter the value of n: "))
        lb = 0
        rb = 100
        arr = [random.randint(lb, rb) for _ in range(n)]
        print()
        best_time = best_merge_sort(n)
        print(f"Time Taken for best case = {best_time * 1e6} microseconds")
        average_time = average_merge_sort(arr.copy())
        print(f"Time Taken for average case = {average_time * 1e6} microseconds")
        worst_time = worst_merge_sort(n)
        print(f"Time Taken for worst case = {worst_time * 1e6} microseconds")
        print()
```

```python
import time
import random

def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, i, 0)

def average_heap_sort(arr):
    start = time.time()
    heap_sort(arr)
    finish = time.time()
    return finish - start

def best_heap_sort(n):
    arr = list(range(n))
    start = time.time()
    heap_sort(arr)
    finish = time.time()
    return finish - start

def worst_heap_sort(n):
    arr = list(range(n, 0, -1))
    start = time.time()
    heap_sort(arr)
    finish = time.time()
    return finish - start

if __name__ == "__main__":
    T = int(input("Enter the number of values of n: "))
    for k in range(T):
        n = int(input("Enter the value of n: "))
        lb = 0
        rb = 100
        arr = [random.randint(lb, rb) for _ in range(n)]
        print()
        best_time = best_heap_sort(n)
        print(f"Time Taken for best case = {best_time * 1e6} microseconds")
        average_time = average_heap_sort(arr.copy())
        print(f"Time Taken for average case = {average_time * 1e6} microseconds")
        worst_time = worst_heap_sort(n)
        print(f"Time Taken for worst case = {worst_time * 1e6} microseconds")
        print()
```

```python
import time
import random

def counting_sort(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10

    for i in range(n):
        index = (arr[i] // exp) % 10
        count[index] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    i = n - 1
    while i >= 0:
        index = (arr[i] // exp) % 10
        output[count[index] - 1] = arr[i]
        count[index] -= 1
        i -= 1

    for i in range(n):
        arr[i] = output[i]

def radix_sort(arr):
    max_val = max(arr)
    exp = 1

    while max_val // exp > 0:
        counting_sort(arr, exp)
        exp *= 10

def average_radix_sort(arr):
    start = time.time()
    radix_sort(arr)
    finish = time.time()
    return finish - start

def best_radix_sort(n):
    arr = list(range(n))
    start = time.time()
    radix_sort(arr)
    finish = time.time()
    return finish - start

def worst_radix_sort(n):
    arr = list(range(n, 0, -1))
    start = time.time()
    radix_sort(arr)
    finish = time.time()
    return finish - start

if __name__ == "__main__":
    T = int(input("Enter the number of values of n: "))
    for k in range(T):
        n = int(input("Enter the value of n: "))
        lb = 0
        rb = 100
        arr = [random.randint(lb, rb) for _ in range(n)]
        print()
        best_time = best_radix_sort(n)
        print(f"Time Taken for best case = {best_time * 1e6} microseconds")
        average_time = average_radix_sort(arr.copy())
        print(f"Time Taken for average case = {average_time * 1e6} microseconds")
        worst_time = worst_radix_sort(n)
        print(f"Time Taken for worst case = {worst_time * 1e6} microseconds")
        print()
```

```python
class Item:
    def __init__(self, weight, value):
        self.weight = weight
        self.value = value
        self.ratio = 0.0

def compare_items(item1, item2):
    return item1.ratio > item2.ratio

def fractional_knapsack(items, capacity):
    for item in items:
        item.ratio = item.value / item.weight
    items.sort(key=lambda x: x.ratio, reverse=True)
    total_value = 0.0
    current_weight = 0

    for item in items:
        if current_weight + item.weight <= capacity:
            total_value += item.value
            current_weight += item.weight
        else:
            remaining_weight = capacity - current_weight
            total_value += (remaining_weight / item.weight) * item.value
            break

    return total_value

if __name__ == "__main__":
    n = int(input("Enter the number of items: "))
    items = []
    print("Enter the weight and value of each item:")
    for _ in range(n):
        weight, value = map(int, input().split())
        items.append(Item(weight, value))

    capacity = int(input("Enter the knapsack capacity: "))
    max_value = fractional_knapsack(items, capacity)
    print("The maximum value that can be obtained is:", max_value)
```

```python
import heapq

class Edge:
    def __init__(self, to, weight):
        self.to = to
        self.weight = weight

class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.adj = [[] for _ in range(vertices)]

    def add_edge(self, from_vertex, to_vertex, weight):
        edge1 = Edge(to_vertex, weight)
        edge2 = Edge(from_vertex, weight)
        self.adj[from_vertex].append(edge1)
        self.adj[to_vertex].append(edge2)

    def prim_mst(self):
        pq = [(0, 0)]
        in_mst = [False] * self.V
        key = [float('inf')] * self.V
        mst_weight = 0
        key[0] = 0

        while pq:
            weight, u = heapq.heappop(pq)

            if in_mst[u]:
                continue

            in_mst[u] = True
            mst_weight += key[u]

            for edge in self.adj[u]:
                v = edge.to
                edge_weight = edge.weight

                if not in_mst[v] and edge_weight < key[v]:
                    key[v] = edge_weight
                    heapq.heappush(pq, (key[v], v))

        return mst_weight

if __name__ == "__main__":
    V, E = map(int, input("Enter the number of vertices and edges: ").split())
    g = Graph(V)

    print("Enter the edges (from, to, weight):")
    for _ in range(E):
        from_vertex, to_vertex, weight = map(int, input().split())
        g.add_edge(from_vertex, to_vertex, weight)

    mst_weight = g.prim_mst()
    print("Minimum Cost Spanning Tree (MST) weight:", mst_weight)
```

```python
class Edge:
    def __init__(self, from_vertex, to_vertex, weight):
        self.from_vertex = from_vertex
        self.to_vertex = to_vertex
        self.weight = weight

class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.edges = []

    def add_edge(self, from_vertex, to_vertex, weight):
        self.edges.append(Edge(from_vertex, to_vertex, weight))

    def find(self, parent, node):
        if parent[node] == -1:
            return node
        return self.find(parent, parent[node])

    def union_op(self, parent, x, y):
        x_root = self.find(parent, x)
        y_root = self.find(parent, y)
        parent[x_root] = y_root

    def kruskal_mst(self):
        self.edges.sort(key=lambda x: x.weight)
        parent = [-1] * self.V
        mst_weight = 0

        for edge in self.edges:
            root_a = self.find(parent, edge.from_vertex)
            root_b = self.find(parent, edge.to_vertex)

            if root_a != root_b:
                mst_weight += edge.weight
                self.union_op(parent, root_a, root_b)

        return mst_weight

if __name__ == "__main__":
    V, E = map(int, input("Enter the number of vertices and edges: ").split())
    g = Graph(V)

    print("Enter the edges (from, to, weight):")
    for _ in range(E):
        from_vertex, to_vertex, weight = map(int, input().split())
        g.add_edge(from_vertex, to_vertex, weight)

    mst_weight = g.kruskal_mst()
    print("Minimum Cost Spanning Tree (MST) weight:", mst_weight)
```

```python
class DijkstraGraph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[float('inf')] * vertices for _ in range(vertices)]

    def add_edge(self, from_vertex, to_vertex, weight):
        self.graph[from_vertex][to_vertex] = weight
        self.graph[to_vertex][from_vertex] = weight

    def dijkstra(self, src):
        dist = [float('inf')] * self.V
        spt_set = [False] * self.V
        dist[src] = 0

        for count in range(self.V - 1):
            u = self.min_distance(dist, spt_set)
            spt_set[u] = True

            for v in range(self.V):
                if not spt_set[v] and self.graph[u][v] and dist[u] != float('inf') and dist[u] + self.graph[u][v] < dist[v]:
                    dist[v] = dist[u] + self.graph[u][v]

        self.print_solution(dist, src)

    def min_distance(self, dist, spt_set):
        min_dist = float('inf')
        min_index = -1

        for v in range(self.V):
            if not spt_set[v] and dist[v] <= min_dist:
                min_dist = dist[v]
                min_index = v

        return min_index

    def print_solution(self, dist, src):
        print(f"Shortest distances from vertex {src} to all other vertices:")
        print("Vertex\tDistance from Source")
        for i in range(self.V):
            print(f"{i}\t{dist[i]}")

if __name__ == "__main__":
    V, E = map(int, input("Enter the number of vertices and edges: ").split())
    graph = DijkstraGraph(V)

    print("Enter the edges (from, to, weight):")
    for i in range(E):
        from_vertex, to_vertex, weight = map(int, input().split())
        graph.add_edge(from_vertex, to_vertex, weight)

    src = int(input("Enter the source vertex: "))
    graph.dijkstra(src)
```

```python
def knapsack(values, weights, capacity):
    n = len(values)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]

    i, j = n, capacity
    selected_items = []

    while i > 0 and j > 0:
        if dp[i][j] != dp[i - 1][j]:
            selected_items.append(i - 1)
            j -= weights[i - 1]
        i -= 1

    print("Selected items for maximum value:")
    for item in selected_items:
        print(f"Item {item} (Value: {values[item]}, Weight: {weights[item]})")

    return dp[n][capacity]

if __name__ == "__main__":
    capacity = int(input("Enter the capacity of the knapsack: "))
    n = int(input("Enter the number of items: "))

    values = []
    weights = []

    print("Enter the values and weights of the items:")
    for i in range(n):
        value, weight = map(int, input().split())
        values.append(value)
        weights.append(weight)

    max_value = knapsack(values, weights, capacity)
    print("Maximum value in the knapsack:", max_value)
```
