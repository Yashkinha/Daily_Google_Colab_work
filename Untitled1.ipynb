```python
import matplotlib.pyplot as plt
import numpy as np
import time

start = time.time()

def bubbleSort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]
                print(arr)

# arr = np.array([65,23,42,73,56,34,76,23,11])

arr = np.array([5,64,79,79,644,7,46,466,3563,3,434,6967,8,498,85,6,598,64,43,9738,41,94,94,6,67,6789,71,94,97,4,])

# arr = np.random.randn(10)
bubbleSort(arr)
print ("Sorted array is:")
for i in range(len(arr)):
    print ("%d" %arr[i])

end = time.time()

print("Time of execution is ", (end-start)*10**3,"ms")

plt.plot(arr)
plt.show()
```
