```python
#Inheritance
class One:
  def Method_One(self):
    print("One-Base class")

class Two(One):   #Single Inheritance
  def Method_Two(self):
    print("Two- Derived class")

O5 = Two()
O5.Method_Two()
O5.Method_One()
```

```python
class One:
  def Method_One(self,a,b):
    print("One-Base class")
    self.A = a
    self.B = b
    print(self.A,self.B)

class Two(One):   #Single Inheritance   :- A, B, Method_One are inherited
  def Method_Two(self):
    print("Two- Derived class")
    print(self.A,self.B)

O5 = Two()
O5.Method_One(100,200)
O5.Method_Two()
```

```python
#Overriding methods - Multiple Inheritance-hybrid
class A:
  def show(self):
    print("A")
class B(A):
  def show(self):
    print("B")
    super().show()
class C(A):
  def show(self):
    print("C")
    super().show()
class E(A):
  def show(self):
    print("E")
    super().show()
class D(B,C,E):  #multiple inheritance
  def show(self):
    print("D")
    #super().show()
    C.show(self)

Ob = D()
Ob.show()
```

```python
#Config Read
from configparser import ConfigParser
config = ConfigParser()
config.read('config.ini')
# Get a list of all sections
print('Sections: %s' % config.sections())
# You can treat it as an iterable and check for keys
# or iterate through them
if 'main_section' in config:
    print('Main section does exist in config.')
for section in config:
    print('Section: %s' % section)
    for key, value in config[section].items():
	    print('Key: %s, Value: %s' % (key, value))
```
