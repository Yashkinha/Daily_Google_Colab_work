```python
#Higher order functions
def sum_all(function, data):   # e.g. sum_all(p,Q)
  result_sum = 0
  for w in data:
    if(function(w)):         # p(x)  or q(x)
      result_sum = result_sum+ w
  return result_sum
Q=[1,3,4,5,6,7,8,9,10,15,20]  #list
p = lambda x: x
q = lambda x : x%2 == 0
r = lambda x : x%3 ==0
s = lambda x : x%4 ==0
print(sum_all(p,Q))
print(sum_all(q,Q))
print(sum_all(r,Q))
print(sum_all(s,Q))
```

```python
#Example
def add(data):
  return data+2
def prod(data):
  return data*2
def main_fun(function1,function2, number_list):
  result_sum=0
  for num in number_list:
    if (num%3==0):
      result_sum=result_sum+function1(num)
    else:
      result_sum=result_sum+function2(num)
  return result_sum

number_list=[1,3,5,6]
print(main_fun(add, prod, number_list)) # functions as parameters
```

```python
#Polynomial Problem F(x) = 5 + x**2+ 4x**3
D5 = {0:5, 2:1, 3:4}
#key-power:value-coefficient
def func1(D,X):
  sum = 0
  for key in D:
    #print (i)
    sum = sum + D[key]*X**key   #key is power here
  print("Sum of Polynolomial is :", sum)

func1(D5,2)
```

```python
#Inheritance
class One:
  def Method_One(self):
    print("One-Base class")

class Two(One):   #Single Inheritance
  def Method_Two(self):
    print("Two- Derived class")

O5 = Two()
O5.Method_Two()
O5.Method_One()
```

```python
class One:
  def Method_One(self,a,b):
    print("One-Base class")
    self.A = a
    self.B = b
    print(self.A,self.B)

class Two(One):   #Single Inheritance   :- A, B, Method_One are inherited
  def Method_Two(self):
    print("Two- Derived class")
    print(self.A,self.B)

O5 = Two()
O5.Method_One(100,200)
O5.Method_Two()
```

```python
#Overriding methods - Multiple Inheritance-hybrid
class A:
  def show(self):
    print("A")
class B(A):
  def show(self):
    print("B")
    super().show()
class C(A):
  def show(self):
    print("C")
    super().show()
class E(A):
  def show(self):
    print("E")
    super().show()
class D(B,C,E):  #multiple inheritance
  def show(self):
    print("D")
    #super().show()
    C.show(self)

Ob = D()
Ob.show()
```

```python
#Program to create a table and insert values into it.
import sqlite3 as lite
import sys
con = lite.connect('TestIT.db')
with con:
    cur = con.cursor()
    cur.execute("DROP TABLE IF EXISTS IT")
    sql1 = "CREATE TABLE IT (FIRST_NAME CHAR(20) NOT NULL, LAST_NAME CHAR(20), AGE INT, SEX CHAR(1), INCOME FLOAT )"
    sql2 = "INSERT INTO IT(FIRST_NAME, LAST_NAME, AGE, SEX, INCOME) VALUES ('Dev', 'Kumar', 21, 'M', 200000)"
    sql3 = "INSERT INTO IT(FIRST_NAME, LAST_NAME, AGE, SEX, INCOME) VALUES ('Tanishq', 'Kumar', 21, 'M', 300000)"
    sql4 = "SELECT * FROM IT"
    cur.execute(sql1)
    cur.execute(sql2)
    cur.execute(sql3)
    cur.execute(sql4)
    rows = cur.fetchall()
    for row in rows:
        print (row)
con.close()
```
