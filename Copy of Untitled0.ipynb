```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset1.csv')
```

```python
string_to_remove = '[4:36 pm, 09/06/2023] Rishabh:'
```

```python
df = df.applymap(lambda x: x.replace(string_to_remove, ""))
```

```python
print(df)
```

```python
existing_column = df.columns[0]
```

```python
new_df = pd.DataFrame(columns=['Date', 'Name', 'Text', 'Timestamp'])
```

```python
new_df['Date'] = '09/06/2023'
new_df['Name'] = 'Rishabh'
new_df['Text'] = '4:36pm'
new_df['Timestamp'] = df[existing_column]
```

```python
print(new_df)
```

```python
new_df.to_csv("n_dataset.csv", index=False)
```
