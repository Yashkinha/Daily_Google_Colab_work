"""# 23 WAP to access protected attribute and method outside the class"""

```python
class yash:
  name = "Yash" #public by default
  _rollno = 218111 # protected attribute

  def _show(self):  #protected method of a class
    print("Name: ",self.name)

  def display(self):  #public method of class
    print("Rollno: ", self._rollno)
    self._show()

ob4 = yash()  #instance of a class
#ob4._show()  #error private method is not accessible outiside the class
ob4.display()  #call a method
```

"""#26 Write a program for following: i)	Create  four classes: Students, Academics, Sports and  Result ii)	Data members:  Students: Name, Rollno, get_details(string, int) Academics: Marks1, Marks2, Marks3, get_marks(float, float, float) Sports: Grades, get_grades(float)  //Grades: 1.0 to 10.0 Result: Total_Marks, Output_result() iii)	Use appropriate inheritance to display the results.  Hint: //Hybrid Inheritance ------class Students{-- } class Academics : public Students   //single inheritance  {--  } class Sports{--   } class Result : public Academics, public  Sports     //Multiple Inheritance {   }

"""

```python
class student:
  def get_details(self,name1,roll1):
    self.name = name1
    self.roll = roll1
class Academics(student): #//single inheritance
  def get_marks(self,m1,m2,m3):
    self.marks1 = m1
    self.marks2 = m2
    self.marks3 = m3
class Sports:
  def get_grades(self,g1):
    self.grade = g1
class Result(Academics,Sports):  #//multiple inheritance
  def output_results(self):
    self.total_marks = self.marks1+self.marks2+self.marks3
    print(self.name, "Roll no: ", self.roll, self.marks1, self.marks2, self.marks3, "Total marks: ", self.total_marks, "Grade: ", self.grade)
ob = Result()  #instance of derived class Result
ob.get_details("Yash",218111)
ob.get_marks(90,91,92)
ob.get_grades('A')
ob.output_results()
```

"""#27 WAP to solve Diamond problem in python. *Hint: Use super(). On constructors in derived classes*"""


```python

class A:
  def show(self):
    print("In A")
class B(A):
  def show(self):
    print("In B")
    super().show()
class C(A):
  def show(self):
    print("In C")
    super().show()
class D(B,C):
  def show(self):
    print("In D")
    super().show()
ob = D()  #instance of derived class D
ob.show()
```
